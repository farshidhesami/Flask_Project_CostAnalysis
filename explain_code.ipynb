-----------------------------------------APP.PY------------------------------------------------

### app.py:

- 

- The from flask import Flask line imports the Flask module, which is a micro web framework for Python.
- render_template is imported to enable rendering of HTML templates.
- request is imported to handle HTTP requests made to the Flask app.
- calculate_discount is imported from the main module, which is a separate file.



- The @app.route('/') decorator specifies that this function will handle requests to the root URL of the Flask app.
- The index function is defined and returns the rendered index.html template.

- The @app.route('/calculate', methods=['POST']) decorator indicates that this function handles POST requests to the /calculate URL.
- The calculate function is defined, which performs the discount calculation and renders the result.html template with the discount and discounted price.
- It retrieves the cost value from the form submitted in the request using request.form['cost'].
- The calculate_discount function is called from the main module to determine the discount percentage based on the cost.
- The discounted price is calculated by subtracting the product of the cost and discount from the cost.
- Finally, the result.html template is rendered with the discount and discounted price as template variables.

- This block ensures that the Flask app is only run when the script is executed directly (not when imported as a module).
- It starts the Flask development server, which listens for requests and serves the Flask app.

----------------------------------MAIN.PY----------------------------------------------------------

## main.py:

- The calculate_discount function takes the cost as input and determines the discount percentage based on the cost.
- If the cost is less than or equal to 1000, it returns a discount of 10% (0.10).
- If the cost is less than or equal to 2000, it returns a discount of 20% (0.20).
- Otherwise, if the cost is greater than 2000, it returns a discount of 30% (0.30).

------------------------------------index.html---------------------------------------------

## index.html:



- This is an HTML template for the home page of the application.
- It contains a form with a text input field (<input type="text" name="cost" id="cost" required>) for the user to enter the cost.

- index.html: 
- This HTML template is rendered for the home page (/). It contains a form where the user can input the cost.
- The form submits the data to the /calculate route using the POST method.
- The cost is entered as a text input with the name cost.
- The form is required, and upon submission, it triggers the calculation of the discount.

## result.html
- 

- result.html: This HTML template is rendered when the discount is calculated. 
- It displays the discount and the discounted price to the user.

- The discount and discounted price are displayed using template variables ({{ discount * 100 }}% and {{ price }}).
- The values of discount and price are passed from the calculate function in app.py when rendering this template.
